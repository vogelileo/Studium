```python
# your code
plt.scatter(df["SkinThickness"], df["BMI"])
m, b = np.polyfit(df["SkinThickness"], df["BMI"], 1)
plt.plot(df["SkinThickness"], m*df["SkinThickness"] + b, color='green', label='Trend line')
plt.xlabel("Skin Thickness")
plt.ylabel("BMI")
plt.show()

correlation = df["SkinThickness"].corr(df["BMI"])
print("Correlation coefficient between SkinThickness and BMI:", correlation)
```

```python
features = [col for col in df.columns if col != "Outcome"]

plt.figure(figsize=(10,6))

colors = plt.cm.tab10.colors  # a set of 10 colors
for i, col in enumerate(features):
    # Add a small jitter to Outcome for better visualization
    y_jitter = df["Outcome"] + np.random.uniform(-0.05, 0.05, size=len(df))
    plt.scatter(df[col], y_jitter, color=colors[i % len(colors)], label=col, alpha=0.7)

plt.xlabel("Feature Value")
plt.ylabel("Outcome (with jitter)")
plt.title("All Features vs Outcome")
plt.legend()
plt.show()
```

```python
features = [col for col in df.columns if col != "Outcome"]

for col in features:
    plt.figure(figsize=(8,3))
    
    # Outcome 0 → y = 0
    plt.scatter(df[df["Outcome"]==0][col], [0]*len(df[df["Outcome"]==0]), 
                color='blue', alpha=0.5, label='Outcome 0')
    
    # Outcome 1 → y = 1
    plt.scatter(df[df["Outcome"]==1][col], [1]*len(df[df["Outcome"]==1]), 
                color='red', alpha=0.5, label='Outcome 1')
    
    plt.xlabel(col)
    plt.yticks([0,1], ['Outcome 0', 'Outcome 1'])
    plt.title(f"{col} distribution by Outcome")
    plt.legend()
    plt.show()
```
